```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# đường dẫn ảnh
my_path = 'C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/conmeo5.jpg'
# đọc ảnh và hiển thị
im = Image.open(my_path)
# display(im)
plt.imshow(im)
```

```python
# vẽ đồ thị cơ bản
x = [7,11,32,16,5] # tọa độ x nằm ngang
y = [15,20,31,42,33] # tọa độ y nằm ngang
plt.plot(x,y, color = "red") # màu sắc
plt.title("Biểu Đồ") # title (tiêu đề)
plt.axis('off') 
plt.show() # hiển thị
```
