```python
# chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='pink')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=132)
plt.title('biểu đồ Histogram(Biểu đồ tần suất) của ảnh')
plt.xlabel('giá trị của pixel')
plt.ylabel('số lượng')
plt.show()
```

```python
# Làm nổi bật đường biên
plt.figure()
plt.pink()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
# Đường dẫn ảnh
my_path = 'C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/conmeo.jpg'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert("L")

# Hiển thị ảnh trên trục tọa độ
plt.imshow(im)

# Tạo một ình vẽ mới
plt.figure()

# Đặt màu sắc là xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, origin="image")

# Đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
