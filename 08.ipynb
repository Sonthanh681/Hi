```python
import matplotlib.pyplot as plt
```

```python
# vẽ biểu đồ: line chart(biểu đồ đường)
x = [3, 5, 7, 9, 10]
y = [11, 15, 17, 20, 14]

x1 = [3, 7, 2, 5, 9]
y1 = [15, 10, 23, 12, 17]

plt.plot(x, y, 'bo-', label='dữ liệu 1') # vẽ đường 1
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu 2')# vẽ đường 2
plt.legend() # hiển thị chú giải
plt.xlabel('Trục x') # đặt nhãn cho trục x
plt.ylabel('Trục y') # đặt nhãn cho trục y
plt.show() # hiển thị biểu đồ
```

```python
# vẽ biểu đồ cột Bar Chart
categories = ['Năm 1', 'Năm 2', 'Năm 3', 'Năm 4', 'Năm 5']
values = [32, 24, 52, 11, 34]

plt.bar(categories, values, color='g', alpha=0.6) # vẽ biểu đồ cột
plt.xlabel('PC') # đặt nhãn cho trục x
plt.ylabel('số lượng bán ra') # đặt nhãn cho trục y
plt.show() # hiển thị biểu đồ
```

```python
# vẽ biểu đồ hình tròn Pie Chart(biểu đồ hình tròn)
my_labels = ['Tháng 1', 'Tháng 2', 'Tháng 3', 'Tháng 4', 'Tháng 5', 'Tháng 6']
sizes = [21, 30, 42, 26, 32, 15]
my_colors = ['red', 'green', 'blue', 'yellow', 'grey', 'pink']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.0f%%') # vẽ biểu đồ hình tròn
plt.title('Tổng lượng PC đã bán trong năm')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [6, 8, 2, 5, 1]
y = [12, 16, 10, 20, 15]

plt.scatter(x, y, c='r', marker='o', label='dữ liệu')  # vẽ biểu đồ phân tán
plt.title('biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('trục x')  # đặt nhãn cho trục x
plt.ylabel('trục y')  # đặt nhãn cho trục y
plt.legend()  # giển thị chú giải
plt.show()  # hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [11, 32, 55, 31, 12, 14, 64, 57, 32]

plt.boxplot(data)  # vẽ biểu đồ hộp
plt.title('biểu đồ hộp ví dụ')  # đặt tiêu đề cho biểu đồ
plt.ylabel('giá trị')  # dặt nhãn cho trục y
plt.show()  # hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [62, 31, 17, 55, 24, 42, 37, 41, 70]

plt.violinplot(data)  # vẽ biểu đồ violin
plt.title('biểu đồ violin ví dụ')  # đặt tiêu đề cho biểu đồ
plt.ylabel('giá trị')  # đặt nhãn cho trục y
plt.show()  # hiển thị biểu đồ
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python là một ngôn ngữ  tuyệt vời để trực quan hóa dữ liệu và đám mây từ."

wordcloud = WordCloud(width=650, height=300).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('biểu đồ đám mây từ văn bản ví dụ')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```
