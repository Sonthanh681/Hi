```python
# import lớp image
from PIL import Image
from imgtools import *
from IPython.display import display # hiện thị ảnh
```

```python
# đường dẫn ảnh
my_path = 'C:/HuynhMinhTien/THIGIACMAYTINH/Hinh/conmeo4.jpg'
```

```python
# đọc ảnh 
img = load_image(my_path)
# resize(thay đổi kích thước ảnh)
# img = img.resize((250,250))
img = img.resize ((150,150))
```

```python
display(img)
```

```python
# chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
